## <span class="motutor-highlight motutor-id_cz27ka4-id_f7i799j"><i></i>60进制 vs 10进制</span>

* 10点30分可以看作是0点开始的10小时又30分，即$10\times60+30=630$分钟
* 同理12点15分是$12\times60+15=735$分钟
* 因此时间差是$735-630=105$分钟
* 105分钟是以10进制表示的时间，再换算回60进制
    * $105\div60=1$小时
    * $105\pmod{60} =45$分钟


## 时间差计算



````python
hour1, minute1 = map(int, input().split())
hour2, minute2 = map(int, input().split())
diff = hour2*60+minute2-(hour1*60+minute1)
print(diff//60, diff%60)
````



* 为什么要用`diff//60`？



## <span class="motutor-highlight motutor-id_utfj9em-id_pohld9m"><i></i>另一种计算方法</span>
```
10 30
12 15
```
的时候，直接计算`diff_minute = minute2 - minute1`会得到负数。
* 如果能否有办法判断`diff_minute`是否是负数，当它是负数的时候，做一点调整，也能得到正确的结果
* 对于上面的值，直接计算的`diff_hour`是2，而`diff_minute`是-15
* 这和10-6的时候，个位要借位是一样的
* 因此，`diff_minute`需要`+60`，而`diff_hour`需要`-1`




## 判断



````python
diff_minute=-15
diff_hour = 2
if diff_minute < 0:
    diff_minute += 60   #相当于diff_minute = diff_minute + 60
    diff_hour -= 1      #相当于diff_hour = diff_hour - 1
print(diff_hour, diff_minute)
````


* <span class="motutor-highlight motutor-id_vu2cem1-id_4kwn8j0"><i></i>`if`</span>是Python的语句，用来根据判断的结果采取不同的行动
* `diff_minute<0`是一个表达式，其中的`<`是一个关系运算符，计算左右两边的关系
* 表达式后面的冒号`:`表示接下去的行是这个`if`的一部分
* 在`if`之后的行，如果保持相同的缩进，就是这个`if`的一部分，是当表达式的结果是`True`，即关系成立的时候，要执行的内容
* `print(diff_hour, diff_minute)`和`if`是对齐的，因此不是`if`的一部分。无论`if`的条件成立与否都要执行
