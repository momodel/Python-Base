
## 变量

* <span class="motutor-highlight motutor-id_3re50wn-id_ubo93a7"><i></i>`shuru = input()`</span>：这里的`shuru`是一个变量
* 当我们需要记录数据（值）的时候，就可以像这样将值赋给变量
* 变量记录了值，是值的名字


```python
a = 6
wendu = 36.2
a = 'hello'
huida = "你好"
print(a)
print(wendu)
print(huida)
```


* *Python的变量不需要事先声明或定义*



## 变量的名字

变量的命名有一些规则：
* 必须以字母、下划线开头，后面可以跟任意数量的字母、数字和下划线。此处的字母并不局限于英文字母，可以包含中文字符、表情符号等
* 变量名区分大小写
* 变量名中不能有空格
* 变量名不能以数字开头
* 不能使用Python语言的关键字作为变量名
* 不要使用内置函数的名字作为变量名，比如`max`




## 输出变量的值

* `print()`可以<span class="motutor-highlight motutor-id_53cl4c6-id_0apjqj2"><i></i>输出变量的值</span>
    * `print(shuru)`
* `shuru`是变量的名字，在需要一个值的地方出现变量的名字，表示我们需要取得那个变量的值
* `print("shuru")`输出的是`shuru`这个字符串的内容
* 理解变量的名字和具体的值的区别是非常关键的基石性的知识



```python
shuru=12
print(shuru)
print('shuru')
```


## 数据类型

* 一个<span class="motutor-highlight motutor-id_ry4p1kr-id_qf84rws"><i></i>变量有两个属性</span>：名字和值
* 而值是具有类型的
    * `a = 6`和`b = "hello"`，两者的值的类型不同
* `type()`告诉我们这个变量的值的类型
    * `int`：整数
    * `str`：字符串

* *和某些其他编程语言不同，Python的变量没有类型*
