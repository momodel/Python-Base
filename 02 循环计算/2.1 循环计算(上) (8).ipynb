## 辗转相除Euclid

变形之后，代码可以进一步简化：



```python
a, b = map(int, input().split())
while True:
    if b==0:
        break
    #r = a%b
    #a = b
    #b = r
    a, b = b, a%b
print(a)
```



* `#`是Python的注释符号，从`#`到行末的内容是给人阅读的，计算机不管
* 当需要临时从程序中去掉一些内容，又不希望直接删除时，可以打上注释




## 辗转相除Euclid

还可以进一步变形：



```python
a, b = map(int, input().split())
#while True:
#    if b==0:
#        break
while b:
    a, b = b, a%b
print(a)
```



* 当`if...break`是`while True`的第一句时，可以将`if`的条件取反后写到`while`后面代替`True`，并去掉`if`语句



## 辗转相除Euclid

于是，整个Eculid算法的核心只有一句：


```python
while b > 0:
    a, b = b, a % b
```



* 有必要先判断x和y的大小，保证进入循环的时候，x是较大的数吗？
* 求GCD是非(zhí)常(dé)基(bèi)本(sòng)的算法，常用于分数约分等场合
* $\frac {12}{18} \rightarrow \frac {2}{3}$
