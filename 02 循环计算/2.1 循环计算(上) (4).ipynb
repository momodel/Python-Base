## 数整数位数

1. 对一个整数做`//10`可以去掉它的个位数字
2. 重复1的步骤，直到这个数字变成0



```python
x = int(input())
cnt = 0
while True:
    if x==0:
        break
    cnt += 1
    x //= 10
print(cnt)
```


## <span class="motutor-highlight motutor-id_vmaptdz-id_8uq3kc8"><i></i>手工运行程序</span>

* 画出变量表格
* 人工执行程序，将变量的每一次变化写在表格上
* 当程序执行结束，表格上每一列最后的值就是程序结果


```python
x = int(input())         #x      cnt
cnt = 0
while True:
    if x==0:
        break
    cnt += 1
    x /= 10
print(cnt)
```
