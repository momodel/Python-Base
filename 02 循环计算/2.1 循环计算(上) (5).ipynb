
## 赋值

* Python的赋值是语句
* 两种特殊的赋值：
    * 连续赋值：`a = b = 6`
    * 同步赋值：`a, b = 4, 5`
    


```python
a, b = 4, 5
print(a, b)
```


## 同步赋值

`a, b = 4, 5`

同步的意思是多个变量的赋值是同时发生的，没有时间先后。
