
## 阶梯电价

> 第一档年累计用电量2760度以内，每度电为0.538元，第二档电量2761—4800度，以每度电0.588元结算；第三档年用电量4800度以上，为每度电0.838元。

* 从文字描述-->数学描述

$$
y = \begin{cases}
0.538x, x\leqslant2760\\
0.588x, x\in[2761, 4800]\\
0.838x, x\gt4800
\end{cases}
$$




## 阶梯电价

$$
y = \begin{cases}
0.538x, x\leqslant2760\\
0.588x, x\in[2761, 4800]\\
0.838x, x\gt4800
\end{cases}
$$
* 从数学描述-->代码描述


```python
if x<=2760:
    y = 0.538 * x
elif x<=4800:   # 为什么这里不需要判断是否>2760
    y = 0.588 * x
else:
    y = 0.838 * x
```


## <span class="motutor-highlight motutor-id_jtr7h9d-id_9lsp0ww"><i></i>连缀的`if...elif...else`</span>


```python
if x<=2760:
    y = 0.538 * x
elif x<=4800:
    y = 0.588 * x
else:
    y = 0.838 * x
```


* `elif`就是`else if`，但是没有`elif`，代码会变成：


```python
if x<=2760:
    y = 0.538 * x
else:
    if x<=4800:
        y = 0.588 * x
    else:
        y = 0.838 * x
```

> 用户可自行选择峰谷分时电价，其中第一档电量峰、谷电价为每度电0.568元、0.288元。而第二、三档电量峰、谷电价在第一档的基础上相应提高5分钱和0.3元
