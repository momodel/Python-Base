## <span class="motutor-highlight motutor-id_jjyi5sh-id_v31tymr"><i></i>列表的函数</span>：`remove()`



```python
a = [2,3,5,7,5,11]
a.remove(5)
a
```


* `remove()`删除找到的第一个元素
* 如果要删除的数据不在列表中，则会发生错误
* `del`是删除指定位置上的元素，而`remove()`是删除指定值的元素
* `[x:x+1]=[]`也可以删除某个位置上的元素
* `[:]=[]`可以清空整个列表




## `pop()`

* 用函数pop()删除并返回列表中指定下标（位置）的数据，如果下标不指定，则删除最后一项



```python
a = [2,3,5,7,11]
print(a.pop())
print(a.pop(2))
a
```


## `reverse()`



```python
a = [2,3,5,7,11]
a.reverse()
a
```


## `index()`



```python
a = [2,3,5,7,11]
a.index(3)
```


* 如果查找的数据在列表中不存在，则会发生错误
* `in`和`index()`都可以用来查找，但是结果和找不到时的表现不同
* 字符串的`find()`函数在列表中不存在，但是字符串也有`index()`函数



```python
a = [2,3,5,7,11]
idx = int(input())
print(a.index(idx))
print(idx in a)
```
