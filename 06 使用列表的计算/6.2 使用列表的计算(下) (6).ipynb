
## <span class="motutor-highlight motutor-id_bs5uy48-id_t5euvcw"><i></i>删除2的倍数</span>



```python
n=int(input())
prime = list(range(2,n))
for k in range(2*2, n, 2):
    prime.remove(k)
print(prime)
```


## 2的下一个是3



```python
n=int(input())
prime = list(range(2,n))
for k in range(3*2, n, 3):
    prime.remove(k)
print(prime)
```


## 2的下一个是x



```python
n=int(input())
prime = list(range(2,n))
for x in prime:
    for k in range(x*x, n, x):
        if k in prime:
            prime.remove(k)
print(prime)
```



* 对于`prime`中的每一个`x`，删掉它的所有的倍数
* 由于$x$的小于$x$的倍数已经在之前出现过了，所以从$x^2$开始
* 加`if k in prime`是因为存在2的倍数也是3的倍数的情况
* `for x in prime`每一轮取下一个元素的时候，是在当时的`prime`列表上取，所以也是动态的，删掉了的数不会再出现




## 小结

* `for`和`while`都可以带`else`，如果没有`break`，就会执行`else`的语句
* `break`只能跳出它所在的那一层循环
