
## 素数

* 只能被1和自己整除的自然数，不包括1
* 围绕素数有很多数学猜想，在理论计算中有大量应用

![](5-1.png)




## 判断素数



```python
import math
x = int(input())
for k in range(2,int(math.sqrt(x)+1)):
    if x%k==0:
        print('not prime')
        break
    else:
        print('prime')
```


## <span class="motutor-highlight motutor-id_fqu9yn6-id_81ltruk"><i></i>又加快一点点</span>

* 所有的偶数都不是素数
* 如果一个数能够被3整除，那么一定能被6、9、12这些3的倍数整除
* 所以，不需要判断$[2,\sqrt x]$的数能否整除，只需要判断$[2,x)$之间所有的素数能否整除
* 这种方法适合构造n以内的素数表，不适合判断单个自然数是否是素数
