
## 数据有<span class="motutor-highlight motutor-id_kc5ghwy-id_2msoqzg"><i></i>确定个数</span>

* 实验数据通常可以先数个个数
`n = int(input())`
* 然后跑一个
`for i in range(n):`
* 这是经典的计数循环



```python
n = int(input())
s = 0
for i in range(n):
    s += int(input())
print(s/n)
```


## 每行有不确定数量的数据

>1 2 3  
4 5  
6 7 8 9 -1

* 首先考虑第一次读入时怎么做



```python
m = list(map(int, input().split()))
lst.extend(m)
if -1 in m:
    lst.remove(-1)
    break
```

## 每行有不确定数量的数据

>1 2 3</br>
4 5</br>
6 7 8 9 -1



```python
lst = []
while True:
    m = list(map(int, input().split()))
    lst.extend(m)
    if -1 in m:
        lst.remove(-1)
        break
```



* 然后用`while True:`包围起来
