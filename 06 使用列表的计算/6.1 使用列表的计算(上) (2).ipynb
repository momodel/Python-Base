


## <span class="motutor-highlight motutor-id_qnax5jq-id_681e42s"><i></i>对元素赋值</span>

* 通过下标可以得到容器的一个元素的值，也可以通过这个方式，修改容器中某个元素的值



```python
t = [1,2,3]
t[1] = 4
t
```



* 但是，字符串是不可修改的数据类型，所以不能修改字符串中的某个字符




## <span class="motutor-highlight motutor-id_ztitb19-id_44lzdlf"><i></i>列表赋值</span>

* <span class="motutor-highlight motutor-id_ztitb19-id_zuuzknq"><i></i>列表变量</span>是列表的管理者而非所有者，列表变量之间的赋值，是管理权的共享，而非列表内容的复制



```python
s1 = [1,2,3]
s2 = s1
s2[0] = 4
print(s1)
print(s2)
```


## <span class="motutor-highlight motutor-id_mj1wbef-id_8gj5txt"><i></i>复制列表</span>

* 用切片可以复制整个列表



```python
s1 = [1,2,3]
s2 = s1[:]
s2[0] = 4
print(s1)
print(s2)
```
