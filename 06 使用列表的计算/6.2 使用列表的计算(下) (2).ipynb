
## 列表和字符串之间的操作

* <span class="motutor-highlight motutor-id_5wfi5cj-id_yw7mgwd"><i></i>字符串`split()`</span>：根据单个字符或字符串分隔字符串为列表，默认为空格
  * 小心当有连续的分隔符时



```python
date = '3/11/2018'
a = date.split('/')
print(a)
name = 'John Johnson'
a = name.split()
print(a)
#多个符号？
```


* 如果split()函数不带参数，就是以空格来分割字符串。




## 数单词

* 求一句英文句子的单词数。单词是字母数字串，中间没有空格



```python
sentence=input()
words=sentence.split()
print(len(words))
```


## <span class="motutor-highlight motutor-id_saagv40-id_axnjq65"><i></i> `join()`</span>

* 字符串`join`：用字符串做分隔符将列表中的元素组成一个字符串



```python
a = ['hello','good','boy','wii']
print(' '.join(a))
print(':'.join(a))
```



* 注意这是字符串的函数




## <span class="motutor-highlight motutor-id_l0p8x1y-id_xr0uykr"><i></i>复制列表</span>



```python
a=[1,2,3,4]
b=a
print(id(a),id(b))
b[2]=5
print(a)
c=a.copy()
print(id(a),id(c))
c[2]=6
print(a)
```



* `id()`给出一个变量的值的编号
