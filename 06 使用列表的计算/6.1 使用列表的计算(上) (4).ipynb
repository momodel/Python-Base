
## <span class="motutor-highlight motutor-id_h81d6de-id_jhgvjn7"><i></i>列表的函数</span>

* 这些函数是列表的函数，是通过`.`运算符在列表上操作的
  * `append()`：在列表后面增加一个元素
  * `extend()`：把另一个列表的内容添加到列表的后面  
  * `insert()`：插入
  * `remove()`：删除第一次出现的那个元素
  * `pop()`：弹出指定位置的元素，不指定则弹出最后一个
  * `reverse()`：反转自己
  * `index()`：查找第一次出现的位置
  * `copy()`：复制列表




## `append()`、`extend()`和`insert()`



```python
a = [2,3,5,7,11]
a.append(13)
print(a)
a = [2,3,5,6,11]
a.extend([13,17])
print(a)
a.append([33,44])
print(a)
a = [2,3,5,6,11]
a.insert(2,4)
a.insert(12,13)  #指定插入的位置(12)不存在，加到最后
print(a)
```


* `append()`和`extend()`添加一个列表时是不同的
* `insert()`插入到指定位置之前，插入的位置不存在时，加到最后
