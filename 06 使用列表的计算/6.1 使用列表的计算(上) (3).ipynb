
## <span class="motutor-highlight motutor-id_5m8x962-id_xhngtds"><i></i>切片赋值</span>

* 列表的切片还可以被赋值。当列表的切片被赋值时，切片所表达的那部分元素被新的列表中的全部元素所替换



```python
name = [1,2,3,4]
name[2:] = [5,6]
name
```


## 切片赋值

* 而且这种替换的元素数量还可以不一致



```python
name = [1,2,3,4]
name[2:] = [5,6,7]
print(name)
name[2:] = [9,10]
print(name)
```



* 那么，当给一个切片赋值一个空列表的时候，会发生什么？



```python
name = [1,2,3,4]
name[2:] = []
print(name)
name[:] = []
print(name)
name[2] = []
print(name)
```
