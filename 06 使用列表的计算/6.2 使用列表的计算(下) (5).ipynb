
## <span class="motutor-highlight motutor-id_lict1b8-id_c6flaw5"><i></i>判断素数</span>



```python
n=int(input())
prime=[]
for x in range(2,n):
    for k in prime:
        if x%k==0:
            break
    else:
        prime.append(x)
print(*prime, sep=',')
```


* `for k in prime`：遍历prime列表，每次取一个已知素数来
* `break`只能跳出它所在的那层循环，所以会继续下一轮的`for x in range(2,n)`
* 如果遍历完了都不能整除，那么x是素数
* `*`把列表打开传递给`print()`
