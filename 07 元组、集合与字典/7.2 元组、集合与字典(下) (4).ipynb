
## <span class="motutor-highlight motutor-id_39sgnc5-id_x3gh4ys"><i></i>月份的英文名字缩写</span>

* 构造`monthname='JANFEBMARAPRMAYJUNJULAUGSEPOCTNOVDEV'
`，那么3月份就是`monthname[6:9]`



```python
monthname='JANFEBMARAPRMAYJUNJULAUGSEPOCTNOVDEV'
m = int(input())
if 1<=m<=12:
    m-=1
    print(monthname[m*3:m*3+3])
```
