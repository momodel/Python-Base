
## <span class="motutor-highlight motutor-id_tap83ow-id_wbpvfe7"><i></i>字典方法</span>

| 方法 | 功效 
|-----|------|
| `keys()` | 返回由全部的键组成的一个序列 |
| `values()` | 返回由全部的值组成的一个序列 |
| `items()` | 返回一个序列，其中的每一项是一个元组，每个元组由键和它对应的值组成 |
| `clear()` | 删除所有条目 |
| `get(key,value)` | 返回这个键所对应的值,如找不到返回`value` |
| `pop(key)` | 返回这个键所对应的值，同时删除这个条目 |




## `get()` vs `[]`

* 函数`get()`和运算符`[ ]`不同之处，在于如果键`key`在字典中不存在，则`get(key)`返回`None`值，而运算符`[ ]`会抛出`KeyError`异常
* 因此在使用`[]`的时候，必须确保`key`的存在，比如放在一个`if key in `里面


```python
score = {'张三':78, '李四':92, '王五':92}
s = input()
print(score.get(s, 0))
if s in score.keys():
    print(score[s])
else:
    print(-1)
print(score[s])
```


## `keys()`、`values()`和`items()`

* 函数`keys()`、`values()`和`items()`都是返回一个迭代器
    * 可以被`for in`遍历
* 由于字典中键不重复，所以`keys()`和`items()`的返回结果可以转换成集合
* 而`values()`返回值由于可能存在重复值，应该转换为列表或元组



```python
score = {'张三':78, '李四':92, '王五':92} 
v = list(score.values())
print(v)
v = set(score.values())
print(v)
```


## 用`items()`实现字典遍历



```python
score = {'张三':78, '李四':92, '王五':89}
for key,value in score.items():
    print(f'{key}:{value}')
```


* `for key,value in score.items():`的意思是，
循环的每一轮，从`score.items()`里拿到一个元组，如
`('张三',78)`，然后相当于做
`key, value = ('张三',78)`
