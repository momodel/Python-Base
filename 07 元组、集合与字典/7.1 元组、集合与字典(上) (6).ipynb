
## <span class="motutor-highlight motutor-id_kizyr5x-id_snbvvo9"><i></i>集合运算</span>

* 通过集合的函数或运算符进行集合的并集、交集、差集和对称差的集合运算
* 假设2个集合：`s1 = {2,3,5,7,11};s2 = {2,3,4,5,6,7}`，下表给出4种集合运算的示例

| 运算 | 函数 | 运算符 | 示例 | 结果 | 说明 |
| ---- | ---- | ---- | ---- | ---- | ---- |
| 并集 | `union()` | $\vert$     | `s1.union(s2)` | `{2,3,4,5,6,7,11}`  | 结果是包含两个集合中所有元素的新集合     |
| 交集 | `intersection()` | `&` | ` s1 & s2` | `{2,3,5,7}` | 交集是只包含两个集合中都有的元素的新集合 |
| 差集 | `difference()` | `-` | `s1 - s2` | `{11}` | s1-s2的结果是出现在s1 但不出现在s2的元素的新集合 |
| 对称差 | `symmertric_difference()` | `^`  | `s1 ^ s2` | `{4,6,11}` | 结果是一个除了共同元素之外的所有元素



```python
s1 = {2,3,5,7,11}
s2 = {2,3,4,5,6,7}
print(s1 | s2)
print(s1 & s2)
print(s1 - s2)
print(s1 ^ s2)
```

## Contains Duplicate

* Given an array of integers, find if the array contains any duplicates. Your program should print true if any value appears at least twice in the array, and it should return false if every element is distinct.



```python
t = list(map(int, input().split()))
s = set(t)
if len(t) == len(s):
    print('no dup')
else:
    print('dup')
```


## 挑选名单

* 要在104位同学中随机挑选20位参加周三上午的连麦分享，但是有7位同学已经在预定的名单中了
* 输入数据：104人名单（学号）和7人名单（学号）
* 输出数据：20人名单和剩下的84人名单



```python
import random
ta = {'3190101849', '3190102191', '3190104143', '3190104515'
, '3190104602', '3190104958', '3190105360'}
md = set()
for i in range(104): 
    md.add(input())
md -= ta
tc = set(random.sample(md, 13))  # 随机取样
tc = tc | ta
print(*tc, sep='\n') 
print('-'*80)
print(*(md-tc), sep=', ')
```
