## <span class="motutor-highlight motutor-id_ct9sz3i-id_odyo2lh"><i></i>元组tuple</span>

* 与列表相似，元组`Tuple`也是个有序序列，用`()`生成。而且元组的字面量也实际上是表达式：



```python
t = (10, 11, 12, 13, 14)
print(t)
a = 6
t = (a, 7, 8, 9, 10)
print(t)
```



* 元组和列表一样可以被索引、切片



```python
t = (10, 11, 12, 13, 14)
print(t[0])
print(t[1:3])
```


## 元组是不可修改的量

* 元组一旦制造出来，就不能修改
    * 不能修改元组的内容，也不能修改元组的元素的值



```python
t = (10, 11, 12, 13, 14)
t[0] = 1
t.append(2)
```


## 单个元素的元组

* `(2)`不能被当作是元组
