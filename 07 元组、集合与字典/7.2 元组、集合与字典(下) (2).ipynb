
## <span class="motutor-highlight motutor-id_8ksatlz-id_t53ynwi"><i></i>字典的基本运算</span>：访问和修改条目

* 直接用`[ ]`运算符，用`<字典> [键]`的形式，访问键所对应的数据



```python
score = {'张三':78, '李四':92}
print(score['张三'])	#访问score中键为'张三'的数据
score['李四'] = 89	#把score中键为'李四'的数据修改为89
score['王五'] = 100 #score中没有键为‘王五’的元素，则增加⼀项，键为’王五’,数据为100。
print(score)
```


## 字典的基本运算：删除条目

* 用del语句，删除指定键的字典条目。


```python
del score['张三']
```


* 如果指定键不存在，则会抛出KeyError异常*




## 字典的基本运算：遍历字典

* 使用遍历循环实现


```python
score = {'张三':78, '李四':92, '王五':89}
for name in score:
    print(f'{name}:{score[name]}')
```


## 字典的基本运算：字典大小

* 用函数`len()`得到字典的条目的数量。


```python
score = {'张三':78, '李四':92, '王五':89}
len(score)
```


## 字典的基本运算：检测

* 用`in`和`not in`运算符检测一个键是否在字典中存在


```python
score = {'张三':78, '李四':92, '王五':89} 
print('李四' in score)
print('小白' in score)
```


## 字典的基本运算：比较

* 用`==`和`!=`比较两个字典是否相同（键和值都相同）


```python
score = {'张三':78, '李四':92, '王五':89}
mark = {'张三':78, '李四':92, '王五':91}
print(score == mark)
```
