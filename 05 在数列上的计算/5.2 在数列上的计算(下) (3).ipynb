
## 确定次数的循环用<span class="motutor-highlight motutor-id_xyc1n2w-id_q1hsp06"><i></i>for</span>



```python
import random

numberOfTrails = int(1e7)  # 科学计数法
numberOfHits = 0

for i in range(numberOfTrails):
    x=random.random()*2-1
    y=random.random()*2-1
    if x*x+y*y<=1:
        numberOfHits+=1

pi = 4*numberOfHits/numberOfTrails
print(pi)
```


## 小结

* 确定次数的循环 —> `for in range`
* 在某种条件下结束的循环 —> `while`
* `while`语句的条件是循环继续的条件
* `break`的条件才是循环结束的条件
