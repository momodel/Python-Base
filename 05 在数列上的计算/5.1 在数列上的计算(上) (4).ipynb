```python
x = int(input())
if x%2==0 and x!=2:
    print('not prime')
else:
    for k in range(3, x, 2):
        if x%k==0:
            print('not prime')
            break
        else:
            print('prime')
```


* 去掉偶数，少判断一半




## 再加快一点点



```python
import math
x = int(input())
for k in range(2,int(math.sqrt(x)+1)):
    if x%k==0:
        print('not prime')
        break
    else:
        print('prime')
```



* `sqrt()`在`math`包内，需要先`import`
* `sqrt()`的结果是浮点数，需要`int()`
* 为了保证平方根本身被计算，需要`+1`
* `math.sqrt(x)`会重复计算吗？
