## 先写循环的第一轮




```python
x = int(input())
guess = x/2
if abs(guess*guess - x)<1e-3:
    pass#break
guess = (guess+x/guess)/2
```



* `pass`是空语句
* 不能用 `guess*guess == x` 因为浮点数的计算结果会有误差




## <span class="motutor-highlight motutor-id_nztnrh8-id_uqf1kom"><i></i>完整的代码</span>



```python
import math
x = float(input())
guess = x/2
while True:
    if abs(guess*guess - x)<1e-6:
        break
    guess = (guess+x/guess)/2
print(guess, math.sqrt(x))
```


## 小结

* 循环结束的条件不是输入的数据而是计算的结果
* 浮点数不能用==直接比较相等
