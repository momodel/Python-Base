
## <span class="motutor-highlight motutor-id_v2ovt66-id_lvf6lqu"><i></i>求n以内所有的素数</span>

* 列出[2,n]所有的自然数，逐一判断是否是素数



```python
n = int(input())
for i in range(2,n+1):
    for k in range(2, i):
        if i%k == 0:
            break
        else:
            print(i, end=',')
print()
```


* 两重`for`循环
  * 外面的`for`循环遍历[2, n]，构造数列来测试
  * 里面的`for`循环遍历[2, i)，构造测试用的数列




## <span class="motutor-highlight motutor-id_v3d3t7x-id_c0b5vx8"><i></i>求前n个素数</span>

* 从2开始测试素数，逐一输出，直到有了n个



```python
n = int(input())
x = 2
while n>0:
    for k in range(2, x):
        if x%k == 0:
           break
    else:
        print(x, end=',')
        n -= 1
    x += 1
```


* 一开始不知道循环的终点，无法构造数列来`for`循环
