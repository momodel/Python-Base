
## <span class="motutor-highlight motutor-id_kogem2i-id_qh1tief"><i></i>其他语言的方法</span>



```python
x = int(input())
isprime = True
for k in range(2,x):
    if x%k==0:
        isprime = False
        break
if isprime:
    print(f'{x} is prime')
else:
    print(f'{x} is compsite')
```



* 由于发现的条件是推翻`x`是素数的假设，所以`isprime`要初始化为`True`
* 这个代码没有处理`x==1`的情况




## <span class="motutor-highlight motutor-id_uvpck4f-id_6mt80qz"><i></i>Python的方法</span>



```python
x = int(input())
for k in range(2,x):
    if x%k==0:
        print(f'{x} is compsite')
        break
else:
    print(f'{x} is prime')
```


* `for`和`while`后面的`else`，表示循环正常结束（*无疾而终*），没有因为其中的`break`而结束
* 可以理解为`break`会跳过`else`
