## <span class="motutor-highlight motutor-id_6jh0go1-id_4p1zppm"><i></i>合数</span>是素数的乘积

* 把一个整数x写成若干个素数相乘的形式
* 先让x被2除，每整除一次打印一个`*2`，直到不能被2整除，然后再把剩下的商被3整除，直到被x整除
* 让x依次被$[2, x]$之间的整数k整除，是否需要判断k是否是素数？




## <span class="motutor-highlight motutor-id_iseo71v-id_lwzprxk"><i></i>合数分解</span>



```python
x = int(input())
print(f'{x}=1', end='')
for k in range(2,x+1):
    while True:
        if x%k != 0:
            break
        print(f'*{k}', end='')
        x //= k
print()  
```


* 这个程序能否输出<span class="motutor-highlight motutor-id_rv1vx3k-id_5psz9ln"><i></i>素数的分解</span>
