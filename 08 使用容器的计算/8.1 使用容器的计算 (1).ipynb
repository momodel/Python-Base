## <span class="motutor-highlight motutor-id_brjzwpn-id_i8jxbsd"><i></i>求列表中两数之和等于指定数的下标</span>

* Given nums = [2, 11, 7, 15], target = 9
* Because nums[0] + nums[1] = 2 + 7 = 9
* print [0, 2].



```python
nums = [2, 6, 11, 15, 7, 8]
target = 9
hm = dict()
for i in range(len(nums)):   
    if nums[i] in hm:
        print(hm[nums[i]],i) 
        break
    hm[target - nums[i]] = i
print([2, 6, 11, 15,7,8])
print(hm)
```
