
## <span class="motutor-highlight motutor-id_m6lfg8t-id_lpdt3ja"><i></i>不同数据类型对象和逻辑值的变换</span>

数据类型 | True | False
-------|-------|------
整数 | 非0，如1 | 0
浮点数 | 非0，如1.0 | 0.0
复数 | 非0，如1j | 0+0j
字符串 | 非空字符串，"hello" | ""
逻辑量 | True | False
列表 | 非空列表 | []
元组 | 非空元组 | ()
集合 | 非空集合 | set()
字典 | 非空字典 | {}

* 在需要逻辑值的地方，如`if`和`while`，都可以直接使用其他类型的值（变量）




## <span class="motutor-highlight motutor-id_hvge8sx-id_33worzp"><i></i>集合和字典的应用</span>

* 集合和字典是非常有用的容器，善用它们往往可以事半功倍

* 输入：(以END结束)
```
学生的学号，姓名
学生的学号，课程，成绩
END
```

* 输出：
```
学号，姓名，课程1，课程2，…，平均分
<学号>，<姓名>，<成绩1>，<成绩2>，…， <平均分>
```



## <span class="motutor-highlight motutor-id_3uk9jbh-id_u9rl66w"><i></i>数据结构</span>

* 字典：
    * studentid，键是学号，值是姓名
    * score，键是课程，值是成绩
    * studentscore，键是学号，值是字典score
    * 集合course，课程
