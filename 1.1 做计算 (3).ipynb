## 输入数字

* <span class="motutor-highlight motutor-id_zozxo8b-id_snc5yv0"><i></i>`input()`得到的输入</span>是字符串
* 而`raduis`显然是一个数字
* 如果需要获得数字，则需要进行转换
    * `radius = float(input())`
* `float()`是一个函数，用来从字符串中解析出带有小数点的数的值
* 这样获得的数据的类型是`float`浮点数
* `int`类型只能表达不带小数部分的整数
    * `int()`可以从字符串中解析出整数


## 计算



```python
radius = float(input())
area = 3.14159 * radius ** 2
area
```


* 赋值号（`=`）的右边是一个<span class="motutor-highlight motutor-id_abt2v72-id_1etc05k"><i></i>**表达式（expression）**</span>
* 表达式是由算子（operand)和运算符（operator）组成的计算的式子
* 算子：
    * 字面量：`3.14159`、`2`
    * 变量：`radius`
    * 函数计算的结果
* 运算符：
    * 乘法`*`
    * 幂次`**`
