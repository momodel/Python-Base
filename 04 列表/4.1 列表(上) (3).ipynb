
## <span class="motutor-highlight motutor-id_2q7q70j-id_7ea5y81"><i></i>列表</span>



```python
s=cnt=0
lst = []  # 制造一个空列表，用于存放数据
while True:
    x = int(input())
    if x == -1:
        break
    lst.append(x) # 把x加入到lst中
    cnt += 1
    s += x
print(s/cnt)
print(lst)
```


## 列表list

* `[1, 2, 3, 4]`就是一个列表
* 列表是一个容器类型，它里面可以存放很多数据，数据可以是不同的类型
* `lst=[]`，创建了一个空列表
* `[]`，列表字面量用`[]`表示，里面的元素之间用逗号`,`分隔
* 用`print()`直接输出整个列表的时候，也会输出两端的`[]`



```python
lst=[1,2,3,'a']
print(lst)
```
