## <span class="motutor-highlight motutor-id_hr2ttsr-id_o96ucyg"><i></i>序列的函数</span>

* Python有一些全局的函数，用来对序列做一些基础运算

* `len()`：给出容器中元素的个数。



```python
t = [1,2,3,4]
print(len(t))
```



* 这样，要得到一个容器中最后那个元素，就有了两种办法：



```python
a = [1,2,3,4]
print(a[len(a)-1])
print(a[-1])
```


## 序列的函数

* `min()`和`max()`计算序列中最小和最大的元素
* `sum()`计算序列中元素之和



```python
print(min([1,2,3,4]))
print(max('二三四五'))
print(sum([1,2,3,4]))
```
