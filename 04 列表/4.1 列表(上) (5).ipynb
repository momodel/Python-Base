
## <span class="motutor-highlight motutor-id_vu3mj7j-id_x6l8xx4"><i></i>连接</span>

* 加号可以用来连接两个序列容器成为一个新的大的容器



```python
s1 = 'Hello'
s2 = 'Bye'
print(s1+s2)
t1 = [1,2,3]
t2 = [5,6]
print(t1+t2)
```


* 和数字的加法一样，对两个序列容器做加法，不会修改那两个容器本身




## 重复

* 一个序列容器乘以一个整数，可以形成一个新的重复若干遍的容器



```python
print('6'*23)
print([1,2]*3)
```



* 同样这个运算也不会修改原本的容器内容。




## 存在

* `in`和`not in`判断数据是否在序列中



```python
print('e' in 'Hello')
print(1 in [1,2,3])
print('el' not in 'Hello')
print([1,2] in [1,2,3])
```



* 当需要判断多个元素的组合时，字符串和列表有所不同。




## 比较关系

* 判断是否相等或大小关系



```python
print('12'>'1')
print([1,2]>[2])
print([1,2]<[2])
print([2,2]>[2])
print('12'=='120')
print('123'>'12')
```



* 比较是从左到右逐个元素（字符）比较的，一旦在某个位置上的元素已经可以比较出结果，就不再比较下去了
* 当两个序列不等长时，比较到多出来的那个位置为止，此时短的序列的元素值以0计算
* 对于字符串中的字符，是按照它们的Unicode编码来比较的
