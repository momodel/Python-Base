* `for..in`用来遍历一个序列
  * 对于字符串，遍历的是每一个字符
  * 对于列表，遍历的是每一个元素



```python
s = 'hello'
t = [1,2,3,'start',5]
for x in s:
    print(x, end=',')
print() # 换行
for x in t:
    print(x, end=',')
```


## 遍历

* W3 7-2：对于给定的正整数N，求它的位数及其各位数字之和



```python
shuzi =input()
i = 0 
s = 0
while i < len(shuzi):
    s = s + int(shuzi[i])
    i = i+1
print(f'{i:.0f} {s:.0f}')
```
