## 计算平均数

* 计算平均数是实验数据处理中最常（jiǎn)见(dān)的操作
$$average =  \frac {\displaystyle \sum _{i=0}^{n} a_i} {n}$$
* 输入数据为很多行，每一行一个非负整数
* 最后一行为`-1`，表示输入结束，这个`-1`不是有效数据的一部分

```
1 
3 
2 
5 
11 
6 
-1
```
* <span class="motutor-highlight motutor-id_39lqb9t-id_xdi7t5v"><i></i>计算平均数不需要记录读到的每一个数据</span>




## 计算平均数



````python
s=cnt=0
while True:
    x = int(input())
    if x == -1:
        break
    cnt += 1
    s += x
print(s/cnt)
````
