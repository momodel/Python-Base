* 如果只是要遍历`range()`产生的数列，并不需要转成列表

$$
s = \displaystyle \sum_{i=1}^n\frac{1}{i}
$$



```python
s = 0
n = int(input())
# t = list(range(1,n+1))
# for i in t:
for i in range(1,n+1):
    s += 1/i
print(s)
```


## 错误使用

* 对一个容器用`range()`产生其所有下标的数列再遍历是错误的



```python
x = 'Zhejiang'
for i in range(len(x)):
    print(x[i])
```
