
## <span class="motutor-highlight motutor-id_e0epbq7-id_2suytnh"><i></i>`range()`</span>

* `range()`用来产生一个数列的迭代器
* `list()`可以把这个迭代器转成列表

```python
r = range(1,7)
print(r)
print(type(r))
t = list(r)
print(t)
```


## `range()`

* 如何得到`[2,4,6,8,10]`



```python
t = list(range(2,11,2))
print(t)
```


## `range()`

`range(b, e, s)`

* b: 数列的起点
* e: 数列的终点，不包含在数列中
* s: 等差数列的步进，$a_{i} = a_{i-1} + s$，当$s<0$时，表示递减



```python
print(list(range(10)))
print(list(range(10,0,-1)))
```



* 只有一个参数的`range(n)`相当于`range(0,n)`
