## <span class="motutor-highlight motutor-id_kqygu45-id_33o2yqu"><i></i>切片slice</span>



```python
s='160102199905080118'
s[6:14]
```



* 在`[]`里的冒号，就形成了切片，它可以从字符串中取出其中的一段。对于字符串，取出的一段还是一个字符串
* 注意这里的6是切片开始的下标，而14是切片结束之后的下一个字符的下标
