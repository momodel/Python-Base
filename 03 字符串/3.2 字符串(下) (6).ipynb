
## <span class="motutor-highlight motutor-id_oahtupx-id_uskuk5n"><i></i>遍历字符串</span>

* 遍历：依次访问数据中的每一个成分
* 遍历字符串：依次访问字符串中的每一个字符
* 访问一个字符：`[]`
* 字符的下标：从0到`len()-1`


```python
s = input()
i = 0
while i<len(s):
    print(s[i])
    i += 1
```


## 遍历循环



```python
s = input()
for x in s:
    print(x)
```



* `for...in`用来遍历一个可以被遍历的数据，在循环的每一轮，取数据中的一项作为变量`x`的值来使用
* 可遍历的数据：字符串、列表、元组、集合，以及迭代器
