
## <span class="motutor-highlight motutor-id_0so78vr-id_3hwq6c4"><i></i>复数Complex Number</span>

我们把形如$z=a+bi$（a,b均为实数）的数称为复数，其中a称为实部，b称为虚部，i称为虚数单位，$i^2=-1$。当z的虚部等于零时，常称z为实数；当z的虚部不等于零时，实部等于零时，常称z为纯虚数。

![](3-1.jpeg)

复数可以由其实部和虚部表示二维平面上的一个点，从而表达一个向量。




## 复数

<span class="motutor-highlight motutor-id_7qevs3h-id_f61ci8d"><i></i>Python支持复数运算</span>，在字面量中用字母j代替i。



```python
c1 = 12 + 0.2j
print("c1Value: ", c1)
print("c1Type", type(c1))

c2 = complex(6, 1.2)
print("c2Value: ", c2)

#对复数进行简单计算
print("c1+c2: ", c1+c2)
print("c1*c2: ", c1*c2)
```



* 设$z1=a+bi，z2=c+di (a、b、c、d\in R)$是任意两个复数，那么它们的积$(a+bi)(c+di)=(ac-bd)+(bc+ad)i$




## <span class="motutor-highlight motutor-id_vw03fgz-id_jvvmwba"><i></i>复数计算</span>



```python
aa=10-12j
print(aa.real)  # 实数部分 10.0  
print(aa.imag)  # 虚数部分 -12.0
print(abs(aa))  # 复数的模
print(aa.conjugate()) # 共轭复数
```


* 两个复数$x+yi$与$x-yi$称为共轭复数，它们的实部相等，虚部互为相反数
* 对于复数$z=a+bi, \vert z \vert = \sqrt[2]{a^2+b^2}$称作复数的模
