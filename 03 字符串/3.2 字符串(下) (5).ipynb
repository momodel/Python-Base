
## <span class="motutor-highlight motutor-id_yb9iseg-id_alot24c"><i></i>原始字符串</span>

* 在一个字符串字面量前加一个字符r，表示这个字符串是原始字符串，其中的\不被当作是转义字符前缀。


```python
print(r'hello\nworld')
```

```python
print('hello\nworld')
```


* 输出：`hello\nworld`
* 相当于`r = 'hello\\nworld'`




## 字符串是<span class="motutor-highlight motutor-id_moxqg9s-id_i0ouhua"><i></i>不可修改</span>的

* 字符串中的数据（字符）是不能修改的。



```python
s='hello'
s[0]='k'       #会得到错误
s ='kello'
```


* 可以通过用新的字符串对变量重新赋值，表示新的字符串。


```python
s='hello'
s='bye'
```


* 这样变量s表示字符串'bye'。




## <span class="motutor-highlight motutor-id_obtrcxm-id_sgxral6"><i></i>字符串函数</span>

* 这些函数是字符串的函数，是要用点运算符.放在字符串后面来使用的。比如：

`s = s.lower()`

* 因为字符串是不可修改的数据类型，所以这些函数都不是修改字符串本身，而是产生一个新的字符串



```python
s='HEllpp'
t=s.lower()
print(s, t)
```


## <span class="motutor-highlight motutor-id_m4zm0f0-id_u1hzs6w"><i></i>更多字符串函数</span>

* `find()` 在字符串中查找子字符串所在的位置，如果找不到就返回-1。当子字符串是单个字符时，就是查找这个字符第一次出现的位置
* `find()` 还有第二和第三个参数，用来表示从哪里开始，到哪里结束
* `count()` 统计子字符串出现的次数，当子字符串里面只有一个字符时，就是统计字符出现的次数
* `strip()`、`rstrip()`和`lstrip()` 去掉字符串两端的空格
* `replace()` 替换子字符串为其他子字符串
* `upper()`、`lower()`、`title()` 调整大小写



```python
s='this is a title'
print(s.title())
```
