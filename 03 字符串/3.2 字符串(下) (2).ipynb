| ---- | ---- | ---- | ---- | ---- |
| 3   |  6    |   .   |  1    |  C    |
| -5     | -4    |   -3   |  -2    |  -1  |

- 最右边（后）的字符的下标也可以用`-1`来表示
- 负的下标从右向左递减，从-1开始
- 所以`'36.1C'[-1]`就是`C`


```python
s='36.1C'
print(float(s[0:-1]))
print(s[-1])
print('67.2F'[-1])
wendu=input()
if wendu[-1]=='C':
    print('C')
```


## <span class="motutor-highlight motutor-id_4gbykxp-id_uzb4rb1"><i></i>温度转换</span>

* 输入：华氏或者摄氏温度值、温度标识
* 输出：将华氏温度转换成摄氏温度，摄氏温度转换成华氏温度
* 计算：$华氏=摄氏\times 1.8+32$



```python
wendu = input()
shuzhi = float(wendu[:-1]) # 当切片从0开始时，0可以省略
if wendu[-1] == 'C' or wendu[-1]=='c':
    print(f'{1.8*shuzhi+32:.2f}')   # 格式字符串的{}里可以做表达式计算
else:
    print(f'{(shuzhi-32)/1.8:.2f}')
```
